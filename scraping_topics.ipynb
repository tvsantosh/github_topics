```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
base_url="https://github.com/topics?page="
```

```python
topic_title=[]
for i in range(1, 7):
    new_url = f"{base_url}{i}"
    response = requests.get(new_url)  
    if response.status_code != 200:
        raise Exception(f"Failed to fetch URL {new_url}. Status code: {response.status_code}")
    doc = BeautifulSoup(response.content, 'html.parser')       
    p_tags = doc.find_all('p', {'class': "f3 lh-condensed mb-0 mt-1 Link--primary"})
    for titles in p_tags:
        topic_title.append(titles.text.strip())

```

```python
 topic_title
```

```python

topic_disc=[]
for i in range(1, 7):
    new_url = f"{base_url}{i}"
    response = requests.get(new_url)  
    if response.status_code != 200:
        raise Exception(f"Failed to fetch URL {new_url}. Status code: {response.status_code}")
    doc = BeautifulSoup(response.content, 'html.parser')       
    disc_tags = doc.find_all('p', {'class': "f5 color-fg-muted mb-0 mt-1"})
    for titles in disc_tags:
        topic_disc.append(titles.text.strip())
```

```python
topic_disc
```

```python

topic_url="https://github.com"
topic_urls=[]
for i in range(1, 7):
    new_url = f"{base_url}{i}"
    response = requests.get(new_url)  
    if response.status_code != 200:
        raise Exception(f"Failed to fetch URL {new_url}. Status code: {response.status_code}")
    doc = BeautifulSoup(response.content, 'html.parser')       
    disc_tags = doc.find_all('a', {'class': "no-underline flex-grow-0"})
    for titles in disc_tags:
        topic_urls.append(topic_url+titles['href'])
```

```python
   topic_urls
```

```python
dist={
    'titles': topic_title,
    'discription':topic_disc,
    'urls':   topic_urls
}
```

```python
df=pd.DataFrame(dist)
```
