```python
def get_topic_page(topic_url):
    # Downlode the page
    response=requests.get(topic_url)
    # Check the response
    if response.status_code!=200:
        raise Exception('Fail to load page {}'.formate(topic_url))

    #parse using Beautiful soup
    topic_doc = BeautifulSoup(response.text,'html.parser')
    return topic_doc
    
# -----------------------------------------------------------------------------------------

def get_repo_info(h3_tag, star_tag):
    a_tags = h3_tag.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href']
    
    stars = parse_star_count(star_tag.text.strip())
    
    return username,repo_name,stars,repo_url


# ----------------------------------------------------------------------------------------

def get_topic_repo(topic_doc):
    
    # Get the h1 tags containing repo title,repo url and username
    repo_tags=topic_doc.find_all('h3',{'class':'f3 color-fg-muted text-normal lh-condensed'})
    
    #get star tags
    star_tags=topic_doc.find_all('span',{'class':'Counter js-social-count'})
    topic_title_tag = topic_doc.find('h1', {'class': 'h1'})
    if topic_title_tag:
        topic_title_tag = topic_title_tag.text.strip()
    else:
        topic_title_tag = None


    topic_repo_dict={
        'title':[],
        'username':[],
        'repo_name':[],
        'stars':[],
        'repo_url':[],
        }

    
    for i in range(len(repo_tags)):
        repo_info=get_repo_info(repo_tags[i],star_tags[i])
        topic_repo_dict['title'].append(topic_title_tag)
        topic_repo_dict['username'].append(repo_info[0])
        topic_repo_dict['repo_name'].append(repo_info[1])
        topic_repo_dict['stars'].append(repo_info[2])
        topic_repo_dict['repo_url'].append(repo_info[3])

    return pd.DataFrame(topic_repo_dict)
```

```python
def get_topic_titles(doc):
    topic_title_tags=doc.find_all('p',{'class':'f3 lh-condensed mb-0 mt-1 Link--primary'})
    topic_titles = [] 
    for topics in topic_title_tags:
        topic_titles.append(topics.text)
    return topic_titles


def get_topic_desc(doc):
    topic_desc_tags=doc.find_all('p',{'class':'f5 color-fg-muted mb-0 mt-1'})
    topic_discriptions=[]
    for tag in discription_tags:
        topic_discriptions.append(tag.text.split())
    return topic_discriptions


def get_topic_urls(doc):
    topic_link_tags=doc.find_all('a',{'class':'no-underline flex-grow-0'})
    topic_urls=[]
    base_url="https://github.com"
    for tag in topic_link_tags:
        topic_urls.append(base_url+tag['href'])
    return topic_urls
    
    
    
    

def scrape_topics():
    topics_url="https://github.com/topics"
    response=requests.get(topics_url)
    if response.status_code!=200:
        raise Exception('Fail to load page {}'.formate(topic_url))
    topic_dict={
        'title':get_topic_titles(doc),
        'description':get_topic_desc(doc),
        'url':get_topic_urls(doc)
    }
    return pd.DataFrame(topic_dict)


def scrap_topic(topic_url,path):
    fname=path+'.csv'
    if os.path.exists(path):
        print(f"{path} is alredy exists")
        return
    topic_df=get_topic_repo(get_topic_page(topic_url))
    topic_df.to_csv(path+'.csv',index=None)
    
    


    
    
```

```python
def scrap_topics_repos():
    print("Scraping list of topics")
    topic_df=scrape_topics()
    os.makedirs('data',exist_ok=True) 
    for index,row in topic_df.iterrows():
        print(f"Scraping tops repositorie for {row['title']}")
        scrap_topic(row['url'],f"data/{row['title']}.csv")


scrap_topics_repos()  
```
